```python
import json
```

```python
with open("argo.json", "r") as f:
    data = json.load(f)
```

```python
data['status']['nodes']
```

```python
for key, value in data['status']['nodes'].items():
    if not "geomad-ls.process-id(" in value["name"]:
        continue
    if value["phase"] == "Succeeded":
        continue
    region_code = None
    try:
        for parameter in value["inputs"]["parameters"]:
            if parameter["name"] == "region-code":
                region_code = parameter["value"]
    except KeyError:
        print('no inputs for {}'.format(key))
    print(region_code)
```
